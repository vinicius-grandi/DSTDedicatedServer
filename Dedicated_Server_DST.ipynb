```python
# @title Default title text
import geopy.distance
import requests
from google.colab import runtime

# Fetch IP information using curl
response = requests.get("https://ipinfo.io")
ip_data = response.json()

# Print the dictionary
server_location = ip_data["loc"].split(',')
test = (server_location[0], server_location[1])
brazil = (-23.5558, -46.6396)
distance = geopy.distance.geodesic(server_location, brazil).km
acceptable_distance = 7200 # @param {type:"number"}

print(ip_data["city"])

if distance > acceptable_distance:
 runtime.unassign()

print(str(distance) + "km")
```

```python
# Content of your Bash script
bash_script = '''#!/bin/bash

install_dir="/content/dst"
cluster_name="MyDediServer"
dontstarve_dir="/content/drive/MyDrive"

function fail()
{
    echo Error: "$@" >&2
    exit 1
}

function check_for_file()
{
    if [ ! -e "$1" ]; then
        fail "Missing file: $1"
    fi
}

cd "$steamcmd_dir" || fail "Missing $steamcmd_dir directory!"

check_for_file "$dontstarve_dir/$cluster_name/cluster.ini"
check_for_file "$dontstarve_dir/$cluster_name/cluster_token.txt"
check_for_file "$dontstarve_dir/$cluster_name/Master/server.ini"
check_for_file "$dontstarve_dir/$cluster_name/Caves/server.ini"

/content/Steam/steamcmd.sh +force_install_dir "$install_dir" +login anonymous +app_update 343050 validate +quit

echo 'ServerModCollectionSetup("3021325282")' > /content/dst/mods/dedicated_server_mods_setup.lua


check_for_file "$install_dir/bin64"

cd "$install_dir/bin64" || fail

run_shared=(./dontstarve_dedicated_server_nullrenderer_x64)
run_shared+=(-console)
run_shared+=(-cluster "$cluster_name")
run_shared+=(-monitor_parent_process $$)
run_shared+=(-persistent_storage_root "/content/drive")
run_shared+=(-conf_dir "MyDrive")

"${run_shared[@]}" -shard Caves  | sed 's/^/Caves:  /' &
"${run_shared[@]}" -shard Master | sed 's/^/Master: /'
'''
!mkdir -p /content/dst/bin/scripts/
# Create a file and write the Bash script content
file_path = '/content/dst/bin/scripts/MyDediServer.sh'
with open(file_path, 'w') as file:
    file.write(bash_script)

# Display file path
file_path
```
